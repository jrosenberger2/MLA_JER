##7)

```python
#Rename columns
df = df.rename(columns={'Date/Time':'date_time', 'Temp (C)':'temp_c', 'Dew Point Temp (C)':'dew_point',
                   'Rel Hum (%)':'relative_humidity', 'Wind Spd (km/h)':'wind_speed', 'Visibility (km)':'visibility',
                   'Stn Press (kPa)':'pressure', 'Weather':'weather'})
df.head()
```

##8)

```python
#Standard statistics
df.describe()
```

##9)

```python
#Add column temp_f
#Celsius to Far
df = df.assign(temp_f=df['temp_c']* 1.8+32)
df.head()
```

##10)

```python
#Rearrange columns so temp_f is before temp_c
df = df.reindex(columns=['date_time', 'temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather'])
df.head()
```

##11)

```python
#Show data types of all columns
df.dtypes
```

##12)

```python
#Convert date_time to date/time column
df['date_time'] = pd.to_datetime(df['date_time'])
#df.head()
df.dtypes
```

##13)

```python
#Average per month for temp_f
#df['month'] = pd.to_datetime(df['date_time']).dt.month
monthly_avg = df.groupby(df['date_time'].dt.month)['temp_f'].mean()
monthly_avg
#type(monthly_avg)
```

##14)

```python
#Create a 20x10 scatterplot that compares temperature and relative humidity
#pd.plotting.scatter_matrix(df['temp_f', 'relative_humidity'], figsize=(20,10))
df.plot(x='temp_f',y='relative_humidity', kind='scatter', figsize=(20,10), title="Temperature vs. Relative Humidity", xlabel='Temperature (F)', ylabel='Relative Humidity')
```

##15)

```python
#Create a 20x10 line chart that shows avg temperature per month
monthly_avg.plot(x='month',y='temp_f', figsize=(20,10), title="Average Temperature per Month", ylabel='Temperature (F)', xlabel='Month')
```
